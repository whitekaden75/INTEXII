```python
import pandas as pd
import sqlite3


conn = sqlite3.connect("movies.db")

df = pd.read_sql_query("SELECT * FROM movies_titles", conn)

df.head()
```

```python
# List all your boolean category columns
category_columns = [
    "Action", "Adventure", "Anime Series International TV Shows",
    "British TV Shows Docuseries International TV Shows", "Children", "Comedies",
    "Comedies Dramas International Movies", "Comedies International Movies",
    "Comedies Romantic Movies", "Crime TV Shows Docuseries", "Documentaries",
    "Documentaries International Movies", "Docuseries", "Dramas",
    "Dramas International Movies", "Dramas Romantic Movies", "Family Movies",
    "Fantasy", "Horror Movies", "International Movies Thrillers",
    "International TV Shows Romantic TV Shows TV Dramas", "Kids' TV",
    "Language TV Shows", "Musicals", "Nature TV", "Reality TV", "Spirituality",
    "TV Action", "TV Comedies", "TV Dramas", "Talk Shows TV Comedies", "Thrillers"
]

# Create a new column 'category' which is a list of all category names where the value is True
df["category"] = df[category_columns].apply(
    lambda row: [col for col in category_columns if row[col]], axis=1
)

# Optionally, turn the list into a comma-separated string
df["category_str"] = df["category"].apply(lambda x: ", ".join(x))




```

```python
df = df.drop(columns=category_columns)

```

```python

df.head()
```

```python
df.rename(columns={"category_str": "genre"}, inplace=True)
df.head()
```

```python
df.to_sql("movies_titles", conn, if_exists="replace", index=False)
```
